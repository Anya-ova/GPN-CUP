```python
def get_optimal_price(volume, costs, competitor_price):
    optimal_price = 0
    max_profit = -math.inf
    
    for i in range(int(competitor_price * 0.8), int(competitor_price * 1.21)):
        profit = (i * volume) - (costs * volume)
        
        if profit > max_profit and i <= competitor_price * 1.2:
            max_profit = profit
            optimal_price = i
    
    return optimal_price
```

```python
products = df_with_cost['product'].unique()
places = df_with_cost['place'].unique()

for product in products:
    for place in places:
        # Фильтрация данных для каждого продукта и города
        subset = df_with_cost[(df_with_cost['product'] == product) & (df_with_cost['place'] == place)]
        
        # Создание SARIMAX модели и обучение
        model = SARIMAX(subset['price'], exog=subset['amount'], order=(1, 0, 1), seasonal_order=(1, 1, 1, 12))
        model_fit = model.fit(disp=False)
        
        exog_forecast = pd.DataFrame({'Объем продаж': [subset['amount'].iloc[-3]]*90})
        # Прогноз на 90 дней вперед
        forecast = model_fit.get_forecast(steps=90, exog=exog_forecast)
        forecast_values = forecast.predicted_mean
        
        last_known_price = subset['price'].iloc[-1]
        competitor_price = subset['compet_price'].iloc[-1]
        volume = subset['amount'].iloc[-1]
        costs = subset['cost'].iloc[-1]
        
        # Прогноз с учетом оптимальной цены
        modified_forecast = []
        current_price = last_known_price
        
        for i in range(90):
            if i % 3 == 0:
                optimal_price = get_optimal_price(volume, costs, competitor_price)
                current_price = min(last_known_price + 1, optimal_price)
            modified_forecast.append(current_price)
        
        # Создание таблицы прогноза
        forecast_table = forecast_table.append(pd.DataFrame({'day_num': range(1, 91),
                                                             'product': [product]*90,
                                                             'place': [place]*90,
                                                             'price': modified_forecast}))
```

```python
forecast_table = pd.DataFrame(columns=['day_num', 'product', 'place', 'price'])

# Уникальные значения продуктов и городов
products = df_with_cost['product'].unique()
places = df_with_cost['place'].unique()

# Цикл по каждому продукту и городу
for product in products:
    for place in places:
        # Фильтрация данных для каждого продукта и города
        subset = df_with_cost[(df_with_cost['product'] == product) & (df_with_cost['place'] == place)]
        
        # Создание SARIMAX модели и обучение
        model = SARIMAX(subset['price'], exog=subset['amount'], order=(1, 0, 1), seasonal_order=(1, 1, 1, 12))
        model_fit = model.fit(disp=False)
        
        # Прогноз на 90 дней вперед
        exog_forecast = pd.DataFrame({'Объем продаж': [subset['amount'].iloc[-3]]*90})
        forecast = model_fit.get_forecast(steps=90, exog=exog_forecast)
        forecast_values = forecast.predicted_mean
        
        # Создание таблицы прогноза
        forecast_table = forecast_table.append(pd.DataFrame({'day_num': range(1, 91),
                                                             'product': [product]*90,
                                                             'place': [place]*90,
                                                             'price': forecast_values}))

```

```python
last_known_prices = df_with_cost.groupby(['product', 'place'])['price'].last()

# Получение последней цены конкурента по каждому продукту в городе
competitor_prices = df_with_cost.groupby(['product', 'place'])['compet_price'].last()

# Определение оптимальной цены, удовлетворяющей условиям
for index, row in forecast_table.iterrows():
    product = row['product']
    place = row['place']
    volume = row[df_with_cost['amount'].iterrows()]
    price = last_known_prices.loc[(product, place)]
    competitor_price = competitor_prices.loc[(product, place)]

    optimal_price = price * (1 + 0.2)  # Максимальная цена = 120% от последней известной цены

    if optimal_price > competitor_price:
        optimal_price = competitor_price * 0.8  # Максимальная цена = 80% от цены конкурента

    forecast_table.loc[index, 'price'] = optimal_price

# Вывод прогнозной таблицы
forecast_table
```
